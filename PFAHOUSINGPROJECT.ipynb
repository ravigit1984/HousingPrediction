Here we can see that LotFrontage column has a significant number of missing values i.e. 17%. Let us see the description of this column.

```python
df['LotFrontage'].describe()
```

Here we can see that mean and median of LotFrontage column are 71.03 and 70 respectively. Also, 75th percentile value is 80 but max value 313. This represents that data has outliers. Now let us see 90th, 95th and 99th percentile values.

```python
df['LotFrontage'].describe(percentiles = [0.9, 0.95, 0.99])
```

in  such a case ideally we must impute the values either after outlier removal. So first we remove the outliers then impute the missing values.

```python
#Removing the Outliers
threshold=3
df = df[ np.abs(df['LotFrontage'] - df['LotFrontage'].mean()) / df['LotFrontage'].std() < threshold]
```

```python
df['LotFrontage'].describe(percentiles = [0.9, 0.95, 0.99])
```

```python
#Imputing the missing values
df['LotFrontage'].fillna(df['LotFrontage'].median(), inplace = True)

#Now lets check the Percentage of missing values in each column again
print(round(100*(df.isnull().sum()/len(df.index)), 2))
```
